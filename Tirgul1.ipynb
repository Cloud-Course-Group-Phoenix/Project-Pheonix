```python
#@title Tirgul 1

import ipywidgets as widgets
from IPython.display import display, clear_output

# student data
firstName = widgets.Text(description="First Name:")
lastName = widgets.Text(description="Last Name:")
email = widgets.Text(description="Email:")
link = widgets.Text(description="Link:")
course = widgets.Text(description="Course:")
favouriteShow = widgets.Text(description="Favourite Show:")

class Student:
    def __init__(self, firstName, lastName, email, course, link):
        self.firstName = firstName
        self.lastName = lastName
        self.email = email
        self.course = course
        self.link = link
        self.favouriteShow = ""

def get_students(filename):
    students = []
    with open(filename, 'r') as file:
        for line in file:
            data = line.strip().split(' ')
            student = Student(data[0], data[1], data[2], data[3], data[4])
            students.append(student)
    return students

students = get_students('Tirgul_1/Students.txt')

def display_student_info():
    student_name = student_dropdown.value
    for student in students:
        if student.firstName == student_name:
            firstName.value = student.firstName
            display(firstName)
            lastName.value = student.lastName
            display(lastName)
            email.value = student.email
            display(email)
            link.value = student.link
            display(link)
            course.value = student.course
            display(course)
            print("")
            favouriteShow.value = student.favouriteShow
            display(favouriteShow)
            save_button = widgets.Button(description="Save")
            display(save_button)
            save_button.on_click(save_button_clicked)
            break

student_dropdown = widgets.Dropdown(
    options=[student.firstName for student in students],
    description='Student Name:',
)

def update_student_in_file(student):
    with open('Tirgul_1/Students.txt', 'r') as file:
        lines = file.readlines()

    with open('Tirgul_1/Students.txt', 'w') as file:
        for line in lines:
            data = line.strip().split(' ')
            if data[0] == student.firstName:
                file.write(f"{student.firstName} {student.lastName} {student.email} {student.course} {student.link} {student.favouriteShow}\n")
            else:
                file.write(line)

def save_button_clicked(b):
    print("Save button clicked!")
    for student in students:
        if student.firstName == student_dropdown.value:
            student.favouriteShow = favouriteShow.value
            update_student_in_file(student)
            break

def on_dropdown_change(change):
    if change['type'] == 'change' and change['name'] == 'value':
        clear_output(wait=True)
        display(student_dropdown)
        print("")
        display_student_info()

student_dropdown.observe(on_dropdown_change)

display(student_dropdown)
```
