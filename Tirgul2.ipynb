```python
# @title Tirgul 2

import ipywidgets as widgets
import json
from IPython.display import display, clear_output

studentsList = []

# Student data input fields
firstName = widgets.Text(description="First Name:")
lastName = widgets.Text(description="Last Name:")
email = widgets.Text(description="Email:")
course = widgets.Text(description="Course:")
link = widgets.Text(description="Link:")
favouriteShow = widgets.Text(description="Favourite Show:")

# Load students from file
def get_students():
    global studentsList
    studentsList.clear()
    with open("students.json", "r") as f:
        studentsList.extend(json.load(f))

# Display student
def display_student_info():
    student_name = student_dropdown.value

    for student in studentsList:
        student_n = student["first_name"] + " " + student["last_name"]

        if student_n == student_name:
            # Update widget values
            firstName.value = student["first_name"]
            lastName.value = student["last_name"]
            email.value = student["email"]
            course.value = ", ".join(student.get("courses", []))
            link.value = student["link"]
            if "favourite_show" in student:
              favouriteShow.value = student["favourite_show"]
            else:
              favouriteShow.value = ""

            # Display fields
            display(firstName, lastName, email, course, link, favouriteShow)

            # Save button
            save_button = widgets.Button(description="Save")
            display(save_button)
            save_button.on_click(save_button_clicked)
            break

# Save student info to JSON file
def save_button_clicked(_):
  student_name = student_dropdown.value
  for student in studentsList:
    student_n = student["first_name"] + " " + student["last_name"]

    if student_n == student_name:
      student["first_name"] = firstName.value
      student["last_name"] = lastName.value
      student["email"] = email.value
      student["link"] = link.value
      student["courses"] = [c.strip() for c in course.value.split(",")]
      student["favourite_show"] = favouriteShow.value

      with open("students.json", "w") as f:
            json.dump(studentsList, f, indent=4)
      print("Student information saved successfully!")
      load_drop_down()
      break

def on_dropdown_change(change):
    if change['type'] == 'change' and change['name'] == 'value':
        load_drop_down()

# Load new student
def load_drop_down():
  clear_output(wait=True)
  display(student_dropdown)
  print("")
  display_student_info()

# Load students before initializing dropdown
get_students()

# Dropdown for selecting a student
student_dropdown = widgets.Dropdown(
    options=[student["first_name"] + " " + student["last_name"] for student in studentsList],
    description='Student Name:',
)

# Attach observer to dropdown
student_dropdown.observe(on_dropdown_change, names='value')

# Display dropdown
display(student_dropdown)

# Load dropdown
load_drop_down()
```
